```python
import json 

from bert_extraction.keyword_extraction import KeywordExtraction
from spelling_corrector.spell_corrector import SpellCorrector

spell = SpellCorrector()
keyword = KeywordExtraction()

sentence = 'About this years rebound program'
corrected_sentence = spell.correct_spellings(sentence)
print(corrected_sentence[0]['generated_text'])
keywords = keyword.extract_important_keywords(corrected_sentence[0]['generated_text'])
print(keywords)
keyword_dict = dict((x, y) for x, y in keywords)
json_object = json.dumps(keyword_dict, indent = 4) 
json_object
```
